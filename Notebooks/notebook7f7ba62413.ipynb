```python
import pandas as pd

# Ruta al archivo CSV en Kaggle
file_path = '/kaggle/input/proyectonc/data_latinoamerica.csv'

# Cargar el archivo CSV con pandas
df = pd.read_csv(file_path)

# Visualizar las primeras 10 filas
df.head(10)

```

```python
# Seleccionar las columnas 20 a 28
columns_20_to_28 = df.iloc[:, 20:28]

# Mostrar las columnas seleccionadas
print(columns_20_to_28)
```

```python
# Verificar si hay valores nulos en las columnas seleccionadas
null_counts = columns_20_to_28.isnull().sum()

# Mostrar el conteo de valores nulos por columna
print("Conteo de valores nulos por columna:")
print(null_counts)

# Verificar si hay valores nulos en alguna de las columnas
if null_counts.any():
    print("\nHay valores nulos en las columnas seleccionadas.")
else:
    print("\nNo hay valores nulos en las columnas seleccionadas.")
```

```python
# Imputación por valor constante (ejemplo: reemplazar NaN con 0)
df_filled = df.fillna(value=0)

# Mostrar las primeras filas del DataFrame después de la imputación
print(df_filled.head())
```
