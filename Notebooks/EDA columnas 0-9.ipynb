```python
# Convertir la columna 'date' a formato datetime
df['date'] = pd.to_datetime(df['date'])

# Agrupar por fecha y obtener sumas diarias de casos confirmados y fallecimientos
daily_data = df.groupby('date')[['new_confirmed', 'new_deceased']].sum().reset_index()

# Visualizar las tendencias temporales
plt.figure(figsize=(15, 5))
plt.plot(daily_data['date'], daily_data['new_confirmed'], label='Nuevos Confirmados')
plt.plot(daily_data['date'], daily_data['new_deceased'], label='Nuevos Fallecidos')
plt.title('Tendencias Temporales de Casos Confirmados y Fallecidos')
plt.xlabel('Fecha')
plt.ylabel('Cantidad')
plt.legend()
plt.show()
```

```python
# Convertir la columna 'date' a formato datetime
df['date'] = pd.to_datetime(df['date'])

# Agrupar por fecha y obtener sumas diarias de casos confirmados y fallecimientos
daily_data = df.groupby('date')[['new_confirmed', 'new_deceased']].sum().reset_index()

# Visualizar las tendencias temporales de casos confirmados
plt.figure(figsize=(15, 5))
plt.plot(daily_data['date'], daily_data['new_confirmed'], label='Nuevos Confirmados', color='blue')
plt.title('Tendencias Temporales de Casos Confirmados')
plt.xlabel('Fecha')
plt.ylabel('Cantidad de Nuevos Casos Confirmados')
plt.legend()
plt.show()

# Visualizar las tendencias temporales de fallecimientos
plt.figure(figsize=(15, 5))
plt.plot(daily_data['date'], daily_data['new_deceased'], label='Nuevos Fallecidos', color='red')
plt.title('Tendencias Temporales de Fallecimientos')
plt.xlabel('Fecha')
plt.ylabel('Cantidad de Nuevos Fallecimientos')
plt.legend()
plt.show()
```

```python
# Convertir la columna 'date' a formato datetime
df['date'] = pd.to_datetime(df['date'])

# Agrupar por fecha y obtener sumas diarias de casos confirmados y fallecimientos acumulados
daily_data = df.groupby('date')[['cumulative_confirmed', 'cumulative_deceased']].sum().reset_index()

# Visualizar las tendencias temporales de casos confirmados acumulados
plt.figure(figsize=(15, 5))
plt.plot(daily_data['date'], daily_data['cumulative_confirmed'], label='Casos Confirmados Acumulados', color='blue')
plt.title('Tendencias Temporales de Casos Confirmados Acumulados')
plt.xlabel('Fecha')
plt.ylabel('Cantidad de Casos Confirmados Acumulados')
plt.legend()
plt.show()

# Visualizar las tendencias temporales de fallecimientos acumulados
plt.figure(figsize=(15, 5))
plt.plot(daily_data['date'], daily_data['cumulative_deceased'], label='Fallecimientos Acumulados', color='red')
plt.title('Tendencias Temporales de Fallecimientos Acumulados')
plt.xlabel('Fecha')
plt.ylabel('Cantidad de Fallecimientos Acumulados')
plt.legend()
plt.show()
```

```python

# Agrupar por fecha y obtener la suma diaria de dosis de vacunas administradas acumuladas
daily_data = df.groupby('date')['cumulative_vaccine_doses_administered'].sum().reset_index()

# Visualizar las tendencias temporales de dosis de vacunas administradas acumuladas
plt.figure(figsize=(15, 5))
plt.plot(daily_data['date'], daily_data['cumulative_vaccine_doses_administered'], label='Dosis de Vacunas Administradas Acumuladas', color='green')
plt.title('Tendencias Temporales de Dosis de Vacunas Administradas Acumuladas')
plt.xlabel('Fecha')
plt.ylabel('Cantidad de Dosis de Vacunas Administradas Acumuladas')
plt.legend()
plt.show()
```
