```python
import pandas as pd
```

```python
dep_peru_df = pd.read_csv('Dep_Peru.csv')
```

```python
# Definir el rango de fechas
start_date = '2020-03-06'
end_date = '2022-02-06'

# Filtrar el DataFrame para mantener las filas dentro del rango de fechas
dep_peru_df = dep_peru_df[(dep_peru_df['date'] >= start_date) & (dep_peru_df['date'] <= end_date)]
```

```python
# Definir el nÃºmero de valores nulos que indica una columna para eliminar
threshold_null_values = 23712

# Verificar si hay columnas con valores iguales a 23712 en todas las filas
columns_to_drop = [col for col in dep_peru_df.columns if dep_peru_df[col].isnull().sum() == threshold_null_values]

# AdemÃ¡s, verificar columnas con valores iguales a 23712
for col in dep_peru_df.columns:
    if (dep_peru_df[col] == threshold_null_values).sum() == dep_peru_df.shape[0]:
        columns_to_drop.append(col)

# Eliminar duplicados de la lista de columnas a eliminar
columns_to_drop = list(set(columns_to_drop))

# Eliminar las columnas identificadas
dep_peru_df.drop(columns=columns_to_drop, inplace=True)

# Mostrar cuÃ¡ntas columnas se han eliminado y cuÃ¡ntas quedan
columns_removed = len(columns_to_drop)
columns_remaining = dep_peru_df.shape[1]

print(f"Se han eliminado {columns_removed} columnas.")
print(f"Quedan {columns_remaining} columnas.")
```

```python
# Contar cuÃ¡ntas filas contienen valores nulos en cada columna
null_counts = dep_peru_df.isnull().sum()

# Filtrar solo las columnas que tienen valores nulos
columns_with_nulls = null_counts[null_counts > 0]

# Mostrar el recuento de filas con valores nulos por columna
print("Columns with null values and their counts:")
print(columns_with_nulls)
```
